```python
df['total_revenue'] = (df['item_price']-df['item_tax']) * df['quantity']
product_revenue = df.groupby('product_name')['total_revenue'].sum().reset_index()
product_revenue = product_revenue.sort_values('total_revenue', ascending=False)
print(product_revenue.head(17).sum())
print(product_revenue.head(90).sum())
print(product_revenue.head(365).sum())
print(product_revenue.sum())
```

```python
#here we can see that the top 17 highest selling products are making nearly 50% of the total revenue
#it is also seen that less than 20% products are making 80% of the revenue(pareto's principle')
#another very important point is that 20%(365) of the total products are making more than 90% of the revenue, this shows that 
#pareto's principle holds true here
```

```python
def get_season(date):
    month = date.month
    if month in [12, 1, 2]:
        return 'Winter'
    elif month in [3, 4, 5]:
        return 'Spring'
    elif month in [6, 7, 8]:
        return 'Summer'
    else:
        return 'Fall'

df['Season'] = df['purchase_date'].apply(get_season)
season_sales = df.groupby('Season')['total_revenue'].sum()
print(season_sales)
plt.figure(figsize=(12, 6))
season_sales.plot(kind='bar')
plt.xlabel('Season')
plt.ylabel('Revenue')
plt.title('Sales by Season')
plt.show()
```

```python
df['DayOfWeek'] = df['purchase_date'].dt.day_name()
sales_by_day = df.groupby('DayOfWeek')['total_revenue'].sum()
sales_by_day = sales_by_day.reindex(['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'])
sales_by_day.plot(kind='bar')
plt.xlabel('Day of the Week')
plt.ylabel('Revenue')
plt.title('Sales by Day of the Week')
plt.show()
```
