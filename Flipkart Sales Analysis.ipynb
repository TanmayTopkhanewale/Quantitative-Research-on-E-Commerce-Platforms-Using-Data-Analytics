```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv(r"C:\Users\tanma\Desktop\socbiz\Grad Project\Flipkart_orders.csv")
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['order_date']=pd.to_datetime(df['order_date'],errors='coerce')
df['order_approval_date']=pd.to_datetime(df['order_approval_date'],errors='coerce')
df['order_cancellation_date']=pd.to_datetime(df['order_cancellation_date'],errors='coerce')
df['order_return_approval_date']=pd.to_datetime(df['order_return_approval_date'],errors='coerce')
```

```python
df['dispatch_after_date']=pd.to_datetime(df['dispatch_after_date'],errors='coerce')
df['dispatch_by_date']=pd.to_datetime(df['dispatch_by_date'],errors='coerce')
df['order_ready_for_dispatch_on_date']=pd.to_datetime(df['order_ready_for_dispatch_on_date'],errors='coerce')
df['dispatched_date']=pd.to_datetime(df['dispatched_date'],errors='coerce')
df['deliver_by_date']=pd.to_datetime(df['deliver_by_date'],errors='coerce')
df['order_delivery_date']=pd.to_datetime(df['order_delivery_date'],errors='coerce')
```

```python
df.info()
```

```python
df.columns
```

```python
df['delivery_sla_breached'].fillna("N")
```

```python
df['approval_to_dispatch_days'] = (df['dispatched_date'] - df['order_approval_date']).dt.days
df['dispatch_to_delivery_days'] = (df['order_delivery_date'] - df['dispatched_date']).dt.days
plt.figure(figsize=(12, 6))
df['approval_to_dispatch_days'].plot(kind='hist', bins=40)
plt.xlabel('Days from Approval to Dispatch')
plt.ylabel('Number of Orders')
plt.title('Order Approval to Dispatch Time Distribution')
plt.show()
plt.figure(figsize=(12, 6))
df['dispatch_to_delivery_days'].plot(kind='hist', bins=40)
plt.xlabel('Days from Dispatch to Delivery')
plt.ylabel('Number of Orders')
plt.title('Dispatch to Delivery Time Distribution')
plt.show()
```

```python
df['delivery_sla_breached']=pd.to_numeric(df['delivery_sla_breached'],errors='coerce')
```

```python

fulfillment_analysis = df.groupby(['fulfilment_source', 'fulfilment_type']).agg({'dispatch_to_delivery_days': 'mean','delivery_sla_breached': 'mean'}).reset_index()
plt.figure(figsize=(12, 6))
sns.barplot(x='fulfilment_source', y='dispatch_to_delivery_days', hue='fulfilment_type', data=fulfillment_analysis)
plt.xlabel('Fulfillment Source')
plt.ylabel('Average Dispatch to Delivery Days')
plt.title('Average Delivery Time by Fulfillment Source and Type')
plt.show()

```

```python
order_status_counts = df['order_item_status'].value_counts()

plt.figure(figsize=(12, 6))
order_status_counts.plot(kind='bar')
plt.xlabel('Order Status')
plt.ylabel('Number of Orders')
plt.title('Distribution of Order Statuses')
plt.show()
```

```python
cancellation_reason_counts = df['return_reason'].value_counts().head(15)

plt.figure(figsize=(12, 6))
cancellation_reason_counts.plot(kind='bar')
plt.xlabel('Cancellation Reason')
plt.ylabel('Number of Orders')
plt.title('Distribution of Cancellation/Return Reasons')
plt.show()
```

```python
logistics_performance = df.groupby('pickup_logistics_partner').agg({'dispatch_to_delivery_days': 'mean','delivery_sla_breached': 'mean'}).reset_index()
print(logistics_performance)
plt.figure(figsize=(12, 6))
sns.barplot(x='pickup_logistics_partner', y='dispatch_to_delivery_days', data=logistics_performance)
plt.xlabel('Logistics Partner')
plt.ylabel('Average Dispatch to Delivery Days')
plt.title('Average Delivery Time by Logistics Partner')
plt.show()
```

```python
order_volume_over_time = df.groupby(df['order_date'].dt.to_period('M'))['order_id'].nunique()
plt.figure(figsize=(12, 6))
order_volume_over_time.plot()
plt.xlabel('Month')
plt.ylabel('Number of Orders')
plt.title('Order Volume Over Time')
plt.show()
```

```python
df['is_cancelled'] = df['order_cancellation_date'].notna()
cancellation_rate_over_time = df.groupby(df['order_date'].dt.to_period('M'))['is_cancelled'].mean()

plt.figure(figsize=(12, 6))
cancellation_rate_over_time.plot()
plt.xlabel('Month')
plt.ylabel('Cancellation Rate')
plt.title('Order Cancellation Rate Over Time')
plt.show()
```

```python

df['month'] = df['order_date'].dt.month
orders_by_month = df.groupby('month')['order_id'].nunique()


plt.figure(figsize=(12, 6))
orders_by_month.plot(kind='bar')
plt.xlabel('Month')
plt.ylabel('Number of Orders')
plt.title('Seasonality in Orders')
plt.show()
```

```python
delivery_time_by_quantity = df.groupby('order_quantity')['dispatch_to_delivery_days'].mean()
plt.figure(figsize=(12, 6))
delivery_time_by_quantity.plot(kind='line')
plt.xlabel('Order Quantity')
plt.ylabel('Average Dispatch to Delivery Days')
plt.title('Impact of Order Quantity on Delivery Times')
plt.show()
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose

monthly_order_counts = df.set_index('order_date')['order_id'].resample('M').nunique()

decomposition = seasonal_decompose(monthly_order_counts, model='additive')

plt.figure(figsize=(12, 10))
decomposition.plot()
plt.show()
```
